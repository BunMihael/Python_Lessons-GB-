# Создаем графики

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Air_Traffic_Cargo_Statistics.csv", delimiter = ",")
dfUS = df[df["GEO Region"] == "US"]
dfCA = df[df["GEO Region"] == "Canada"]
dfIN = df[df["GEO Region"] == "Asia"]
```

# Результат

```python
fig, ax = plt.subplots()
plt.plot(pd.to_datetime(dfUS["Activity_Period"], format = "%Y%m" ), dfUS["Cargo_Weight_LBS"], "-*k")
plt.plot(pd.to_datetime(dfCA["Activity_Period"], format = "%Y%m" ), dfCA["Cargo_Weight_LBS"], "-or" )
plt.plot(pd.to_datetime(dfIN["Activity_Period"], format = "%Y%m" ), dfIN["Cargo_Weight_LBS"], "-y")
plt.xlabel("Years", fontname = "Arial", fontsize =15)
plt.ylabel("LBS", fontname = "Arial", fontsize =15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.tight_layout()
plt.show()
```
